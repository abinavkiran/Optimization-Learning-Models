```python
import matplotlib.pyplot as plt
import numpy as np

def fibonacci_search_with_visualization(function, a, b, epsilon, delta):
    """
    Finds the minimum of a unimodal function within a given interval using Fibonacci Search and visualizes each iteration.

    Args:
        function: The unimodal function to be minimized.
        a: The lower bound of the interval.
        b: The upper bound of the interval.
        epsilon: The desired accuracy.
        delta: The step size for final bracketing.

    Returns:
        A tuple with the final interval [a, b] and the approximate minimum of the function.
    """

    def plot_iteration(function, a, b, x1, x2, iteration, f1, f2):
        """
        Visualizes the current iteration of the Fibonacci search.
        """
        x = np.linspace(a_start, b_start, 500)
        y = function(x)

        plt.figure(figsize=(8, 6))
        plt.plot(x, y, label='Function', color='blue')
        plt.axvline(a, color='red', linestyle='--', label='Interval start')
        plt.axvline(b, color='red', linestyle='--', label='Interval end')

        # Highlight the current interval
        plt.fill_between(x, y.min() - 1, y.max() + 1, where=(x >= a) & (x <= b), color='pink', alpha=0.3)

        # Mark the function values at x1 and x2
        plt.scatter([x1, x2], [f1, f2], color='green', label='Function values', zorder=5)
        plt.text(x1, f1, f'f(x1)={f1:.4f}', color='green')
        plt.text(x2, f2, f'f(x2)={f2:.4f}', color='green')

        plt.title(f"Fibonacci Search - Iteration {iteration}")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.legend()
        plt.grid()
        plt.show()

    # Start of the method
    L1 = b - a
    a_start, b_start = a, b  # Save the original interval for plotting
    N = 0
    while Fibonacci(N) < 1 / epsilon:
        N += 1

    x1 = a + (Fibonacci(N - 2) / Fibonacci(N)) * L1
    x2 = b - (Fibonacci(N - 2) / Fibonacci(N)) * L1
    f1 = function(x1)
    f2 = function(x2)

    # Collect iterations data
    iterations_data = []
    iterations_data.append((1, a, b, x1, x2, f1, f2))

    plot_iteration(function, a, b, x1, x2, 1, f1, f2)

    for i in range(2, N):
        Li = (Fibonacci(N - (i - 1)) / Fibonacci(N)) * L1

        if f1 <= f2:
            b = x2
            x2 = x1
            x1 = a + (Fibonacci(N - (i + 1)) / (Fibonacci(N - (i - 1)))) * Li
            f2 = f1
            f1 = function(x1)
        else:
            a = x1
            x1 = x2
            x2 = b - (Fibonacci(N - (i + 1)) / Fibonacci(N - (i - 1))) * Li
            f1 = f2
            f2 = function(x2)

        # Handle the case where f1 == f2
        if abs(f1 - f2) < 1e-8:
            x1 -= 0.001
            f1 = function(x1)

        iterations_data.append((i, a, b, x1, x2, f1, f2))
        plot_iteration(function, a, b, x1, x2, i, f1, f2)

    # Final adjustment with delta
    x1_final = x1 - delta / 2
    x2_final = x2 + delta / 2
    f1_final = function(x1_final)
    f2_final = function(x2_final)

    if f1_final <= f2_final:
        result_point = x1_final
        result_value = f1_final
    else:
        result_point = x2_final
        result_value = f2_final

    # Print the table after all plots
    print(f"\n{'k':<5}{'a':<10}{'b':<10}{'x1':<10}{'x2':<10}{'f1':<10}{'f2':<10}")
    print(f"{'-'*55}")
    for data in iterations_data:
        print(f"{data[0]:<5}{data[1]:.2f}     {data[2]:.2f}     {data[3]:.2f}     {data[4]:.2f}     {data[5]:.2f}     {data[6]:.2f}")

    print(f"\nFinal interval: [{a:.4f}, {b:.4f}]")
    print(f"Approximate minimum point: {result_point:.4f}")
    print(f"Function value at minimum: {result_value:.4f}")

    return (a, b, result_point)

def Fibonacci(n):  # Define the Fibonacci function here
    """
    Calculates the nth Fibonacci number.
    """
    if n <= 1:
        return n
    else:
        return Fibonacci(n-1) + Fibonacci(n-2)

# Example usage with a simple function
def example_function(x):
    return (x-1)*(x-2)*(x-3)

result = fibonacci_search_with_visualization(example_function, 1, 3, 0.13, 0.01)
```
